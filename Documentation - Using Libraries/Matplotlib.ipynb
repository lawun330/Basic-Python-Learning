# Documentation - Using Libraries: Matplotlib
This notebook illustrates the use of the `matplotlib` module.

```python
import matplotlib.pyplot as plt
import numpy as np
from random import sample
```

## TRADITIONAL APPROACH: Using subplot()

- FUNCTION `subplot()` $\rightarrow$ create subplot
- FUNCTION `plot()` $\rightarrow$ plot data
    - parameters: (row, column, index)

- FUNCTION `xlabel()` $\rightarrow$ set label for X-axis
- FUNCTION `ylabel()` $\rightarrow$ set label for Y-axis

- FUNCTION `xticks()` $\rightarrow$ set marking intervals for X-axis
	- empty list, [ ], as parameter $\rightarrow$ remove marking intervals
- FUNCTION `ytick()` $\rightarrow$ set marking intervals for Y-axis
	- empty list, [ ], as parameter $\rightarrow$ remove marking intervals

- FUNCTION `show()` $\rightarrow$ display plot

```python
x = np.linspace(0, 10, 10)
y = x ** 2

plt.subplot(2, 2, 1)
plt.plot(x, y, 'b--') # blue colored dash
plt.xlabel('X-axis name')
plt.ylabel('Y-axis name')
plt.title('Title: y=x^2')

plt.subplot(2, 2, 2)
plt.plot(x, y, 'r+') # red colored plus-sign
plt.xlabel('X-axis name')
plt.ylabel('Y-axis name')
plt.xticks([]) # remove marks on axis
plt.yticks([]) # remove marks on axis

plt.subplot(2, 2, 3)
plt.plot(x, y, 'g*') # green colored asterisks
plt.xlabel('X-axis name')
plt.ylabel('Y-axis name')

plt.subplot(2, 2, 4)
plt.plot(x, y, 'yo-') # yellow colored O-and-dash
plt.xlabel('X-axis name')
plt.ylabel('Y-axis name')
plt.xticks([]) # remove marks on axis
plt.yticks([]) # remove marks on axis

plt.show()
```

## FORMAL OOP APPROACH: Using figure()  with Figure & Axes

**Note: In the following subplot() approach, we do not need to add axes. In here, we need to add them.**

- FUNCTION `figure()` $\rightarrow$ create a blank figure
- FUNCTION `add_axes()` $\rightarrow$ specify [ move_to_right, move_to_top, width, height ] for an axe
    - move_to_right: As this value increases, the axe shifts to the right direction.
    - move_to_top: As this value increases, the axe shifts to the top direction.

- FUNCTION `plot()` $\rightarrow$ plot data

- FUNCTION `set_xlabel()` $\rightarrow$ set label for X-axis
- FUNCTION `set_ylabel()` $\rightarrow$ set label for Y-axis
- FUNCTION `set_title()` $\rightarrow$ set title

- FUNCTION `legend(loc)` $\rightarrow$ display legend in a specified location
    - loc=0 $\rightarrow$ default (location chosen automatically)
    - loc=1 $\rightarrow$ upper right corner
    - loc=2 $\rightarrow$ upper left corner
    - loc=3 $\rightarrow$ lower left corner
    - loc=4 $\rightarrow$ lower right corner

- FUNCTION `show()` $\rightarrow$ display plot
- FUNCTION `savefig()` $\rightarrow$ save the figure in a specified format and dpi (pixels per inch)

```python
# figure() approach
fig = plt.figure()

# adjusting axes
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.4, 0.3, 0.4])

# axes
axes.plot(x, y, 'b', label='y = x**2')
axes.set_xlabel('X-axis name') # notice the use of 'set_' to begin methods
axes.set_ylabel('Y-axis name')
axes.set_title('Axes Title')
axes.legend(loc=4)

# axes 2
axes2.plot(y, x, 'r', label='y = sqrt(x)')
axes2.set_xlabel('X-axis name2')
axes2.set_ylabel('Y-axis name2')
axes2.set_title('Axes 2 Title')
axes2.legend()

plt.show(fig) 

# fig.savefig('filename.png', dpi=200)
```

Note that the two plots are not put within each other. Yet they appear to be in that way.

In reality, there is nothing more than specifiying position coordinates in a canvas.

## COMBINING BOTH APPROACHES: Using subplots() with Figure & Axes

**Note: In the above figure() approach, we need to add axes. In here, we do not need to add them.**


```python
# subplot() method as figure() alternative
fig, axes = plt.subplots() # no parameter as in figure()

axes.plot(x, y, 'g', label='y = x**2')
axes.set_xlabel('X-axis name')
axes.set_ylabel('Y-axis name')
axes.set_title('Title')
axes.legend(loc=2)

plt.show(fig)
```

## INDEXING Axes
- FUNCTION `axis()` $\rightarrow$ get or set some axis properties (can pass different types of arguments)
- FUNCTION `set_xlim([start_index, ending_index])` $\rightarrow$ set marking intervals for X-axis
- FUNCTION `set_ylim([start_index, ending_index])` $\rightarrow$ set marking intervals for Y-axis

Axes can be indexed just like indexing elements in a list.

```python
fig, axes = plt.subplots(1, 3, figsize=(12, 4))

axes[0].plot(x, x**2, x, x**3)
axes[0].set_title("default axes interval")

axes[1].plot(x, x**2, x, x**3)
axes[1].axis('tight') # set limits just large enough to show all data, then disable further autoscaling
axes[1].set_title("tight axes interval")

axes[2].plot(x, x**2, x, x**3)
axes[2].set_xlim([2, 5])
axes[2].set_ylim([0, 60])
axes[2].set_title("custom axes interval")

plt.show(fig)
```

## DIFFERENCE BETWEEN ROW AND COLUMN
The first parameter of the function `subplots()` is the row.
The second parameter is the column. The argument position matters.

```python
# plot with a loop 
# passing (1, 2) creates 1 row 2 columns
fig, axes = plt.subplots(1, 2)

for ax in axes:
    ax.plot(x, y, 'p')
    ax.set_xlabel('X-axis name')
    ax.set_ylabel('Y-axis name')
    ax.set_title('Title')

plt.show(fig)
```

```python
# plot with a loop 
# passing (2, 1) creates 2 rows 1 column
fig, axes = plt.subplots(2, 1)

for ax in axes:
    ax.plot(x, y, 'p')
    ax.set_xlabel('X-axis name')
    ax.set_ylabel('Y-axis name')
    ax.set_title('Title')

plt.show(fig)
```

## EXPLORING OTHER FUNCTIONALITIES with Figure & Axes

- FUNCTION `figure()` $\rightarrow$ create a blank figure
    - `figsize` parameter $\rightarrow$ width and height of a figure
    - `dpi` parameter $\rightarrow$ dots per inch (pixels per inch)

- FUNCTION `plot()` $\rightarrow$ plot data
    - `color` parameter $\rightarrow$ specify color
    - `alpha` parameter $\rightarrow$ specify opacity
        - The default value is 1.
        - As this value decreases, the transparency increases (or the opacity decreases).

```python
# plot with custom size

fig = plt.figure(figsize=(8, 4), dpi=100)

ax  = fig.add_axes([0.1, 0.1, 0.8, 0.8])

ax.plot(x, x**2, color='#8B008B', alpha=0.5)
```

## EXPLORING LINES

- `linewidth` or `lw` parameter $\rightarrow$ specify a line's width
- `linestyle` or `ls` parameter $\rightarrow$ specify a linestyle option
    - Options:
        - -
        - 
        - -.
        - :
        - steps
- FUNCTION `set_dashes()` $\rightarrow$ draw a customized dash line 
    - [ parameters ] $\rightarrow$ [ line length, space length, ... ]
- `marker` parameter $\rightarrow$ specify a marker shape (can be specified along with the `color` parameter)
    - `color="red", marker=+` is the same as `r+` 
    - Options:
        - '+' $\rightarrow$ a plus-sign symbol
        - 'o' $\rightarrow$ a big dot
        - '*' $\rightarrow$ a star
        - 's' $\rightarrow$ a square
        - '.' $\rightarrow$ a small dot
        - '1' or '2' or '3' or 'number...' $\rightarrow$ a concurrent-lines symbol at different rotations
- `markersize` parameter $\rightarrow$ specify a marker's size
- `markerfacecolor` parameter $\rightarrow$ specify a marker's color
- `markeredgewidth` parameter $\rightarrow$ specify the width of a marker's edge
- `markeredgecolor` parameter $\rightarrow$ specify the color of a marker's edge

```python
fig, ax = plt.subplots(figsize=(12,6))

# different linewidths
ax.plot(x, x+1, color="red", linewidth=0.25)
ax.plot(x, x+2, color="red", linewidth=0.50)
ax.plot(x, x+3, color="red", linewidth=1.00)
ax.plot(x, x+4, color="red", linewidth=2.00)

# different linestype options -, , -., :, steps
ax.plot(x, x+5, color="green", lw=3, linestyle='-')
ax.plot(x, x+6, color="green", lw=3, ls='-.')
ax.plot(x, x+7, color="green", lw=3, ls=':')

# custom dash line
line, = ax.plot(x, x+8, color="black", lw=1.50)
line.set_dashes([5, 10, 15]) # format: line length, space length, ...

# different marker symbols
ax.plot(x, x+ 9, color="blue", lw=1, ls='--', marker='+')
ax.plot(x, x+10, color="blue", lw=1, ls='--', marker='o')
ax.plot(x, x+11, color="blue", lw=1, ls='--', marker='*')
ax.plot(x, x+12, color="blue", lw=1, ls='--', marker='s')
ax.plot(x, x+13, color="blue", lw=1, ls='--', marker='.')
ax.plot(x, x+14, color="blue", lw=1, ls='--', marker='1')
ax.plot(x, x+15, color="blue", lw=1, ls='--', marker='2')
ax.plot(x, x+16, color="blue", lw=1, ls='--', marker='3')
ax.plot(x, x+17, color="blue", lw=1, ls='--', marker='4')

# marker size, marker color, edge width, edge color
ax.plot(x, x+18, color="purple", lw=1, ls='-', marker='o', markersize=2)
ax.plot(x, x+19, color="purple", lw=1, ls='-', marker='o', markersize=4)
ax.plot(x, x+20, color="purple", lw=1, ls='-', marker='o', markersize=8, 
        markerfacecolor="red")
ax.plot(x, x+21, color="purple", lw=1, ls='-', marker='s', markersize=8,
        markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green")

plt.show(fig)
```

## RECTANGULAR BOX PLOT
- `vert` parameter
    - True $\rightarrow$ verticle box plot
    - False $\rightarrow$ horizontal box plot
- `patch_artist`
    - True $\rightarrow$ color filled
    - False $\rightarrow$ color not filled

```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

plt.boxplot(data, vert=False, patch_artist=False)
plt.show()
```

```python
plt.boxplot(data, vert=True, patch_artist=True)
plt.show()
```
